```python
import os
from pyspark.sql import SparkSession
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
aws_access_key = os.environ.get("AWS_ACCESS_KEY_ID")
aws_secret_key = os.environ.get("AWS_SECRET_ACCESS_KEY")
print(aws_access_key)
```

```python
spark = (
    SparkSession.builder
    .appName("read parquet")
    .config("spark.jars.ivy", "/tmp/.ivy2")
    .config("spark.jars.packages", "org.apache.hadoop:hadoop-aws:3.4.2")
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem")
    .config("spark.hadoop.fs.s3a.access.key", aws_access_key)
    .config("spark.hadoop.fs.s3a.secret.key", aws_secret_key)
    .getOrCreate()
)
```

```python
df = spark.read.parquet("s3a://github-analytics90416-669003566676-ap-southeast-2-an/parquet1/2026/Jun/28/03")
```

```python
df.printSchema()
```

```python
df.select("created_at").show(10)
```

```python
from pyspark.sql import functions as f
```

```python
df.select("type").distinct().show()
```

```python
from pyspark.sql.types import StructType, StructField, StringType, LongType, DateType, BooleanType, IntegerType
```

```python
df1 = df.filter(f.col("type")=="IssuesEvent").show(n=1, truncate=False, vertical=True)
```

```python
issues_schema = StructType([
    StructField("action", StringType()),
    StructField("issue", StructType([
        StructField("id", LongType()),
        StructField("number", IntegerType()),
        StructField("created_at", DateType()),
        StructField("closed_at", DateType()),
        StructField("state", StringType())
    ]))
])
```

```python
updated_df = df.withColumn("issue_payload", f.when(f.col("type")=="IssuesEvent", f.from_json("payload", issues_schema)))
```

```python
updated_df.printSchema()
```

```python
updated_df = updated_df.withColumn("issue_action", f.col("issue_payload.action")).withColumn("issue_id", f.col("issue_payload.issue.id"))\
                       .withColumn("issue_number", f.col("issue_payload.issue.number")).withColumn("issue_created_at", f.col("issue_payload.issue.created_at"))\
                       .withColumn("issue_closed_at", f.col("issue_payload.issue.closed_at")).withColumn("issue_state", f.col("issue_payload.issue.state"))
```

```python
updated_df=updated_df.drop("issue_payload")
```

```python
updated_df.printSchema()
```

```python
updated_df.filter(f.col("type") == "IssuesEvent").show(n=1,vertical=True,truncate=False)
```

```python
df.filter(f.col("type")=="PushEvent").show(n=2,truncate=False,vertical=True)
```

```python
pushSchema = StructType([
    StructField("push_id", LongType()),
    StructField("repository_id", StringType()),
    StructField("ref", StringType())
])
```

```python
updated_df = updated_df.withColumn("push_payload", f.when(f.col("type")=="PushEvent", f.from_json("payload", pushSchema)))
```

```python
updated_df = updated_df.withColumn("push_id", f.col("push_payload.push_id"))\
    .withColumn("push_repo_id", f.col("push_payload.repository_id"))\
    .withColumn("push_ref", f.col("push_payload.ref"))
```

```python
updated_df = updated_df.drop("push_payload")
```

```python
updated_df.printSchema()
```

```python
updated_df.filter(f.col("type")=="PushEvent").show(n=1,truncate=False,vertical=True)
```

```python
updated_df.filter(f.col("type")=="PullRequestEvent").show(n=1,truncate=False,vertical=True)
```

```python
prSchema = StructType([
    StructField("action", StringType()),
    StructField("number", IntegerType()),
    StructField("pull_request", StructType([
        StructField("id", LongType())
    ]))
])
```

```python
updated_df = updated_df.withColumn("pr_payload", f.when(f.col("type") == "PullRequestEvent", f.from_json("payload", prSchema)))\
    .withColumn("pr_repo", f.when(f.col("type") == "PullRequestEvent", f.col("repo.name")))
```

```python
updated_df = updated_df.withColumn("pr_action", f.col("pr_payload.action")).withColumn("pr_number", f.col("pr_payload.number"))\
              .withColumn("pr_id", f.col("pr_payload.pull_request.id"))
```

```python
updated_df.filter(f.col("type")=="PullRequestEvent").show(n=1,truncate=False,vertical=True)
```

```python
updated_df.filter(f.col("type")=="ForkEvent").show(n=1,truncate=False,vertical=True)
```

```python
updated_df = updated_df.withColumn("is_fork", f.when(f.col("type") == "ForkEvent", f.lit("Y")))\
    .withColumn("fork_repo", f.when(f.col("type") == "ForkEvent", f.col("repo.name")))
```

```python
updated_df.filter(f.col("type")=="ForkEvent").show(n=1,truncate=False,vertical=True)
```
